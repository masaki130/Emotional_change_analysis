```python
# Lについて
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 設定
sp = "L"
bins = 50

# 入力ファイル
input_csv = f"/home/mitani/CSJ-emo-int_bunseki/0718/0720/LR_bunseki/data/vad_delta_{sp}.csv"

# CSV読み込み
df = pd.read_csv(input_csv)

# 各ヒストグラムの頻度を取得
arousal = df["delta_arousal"].dropna().to_numpy()
int_kushinada = df["delta_int_kushinada"].dropna().to_numpy()
hist1, _ = np.histogram(arousal, bins=bins)
hist2, _ = np.histogram(int_kushinada, bins=bins)

# 最大頻度を取得
ymax = max(hist1.max(), hist2.max())

# ヒストグラム
plt.figure(figsize=(6,4))
plt.hist(df["delta_arousal"], bins=bins, edgecolor="black")
plt.title("Delta Arousal L")
plt.xlabel("Δa")
plt.ylabel("Frequency")
plt.ylim(0, ymax)
plt.show()

plt.figure(figsize=(6,4))
plt.hist(df["delta_int_kushinada"], bins=bins, edgecolor="black")
plt.title("Delta Intencity L")
plt.xlabel("Δint")
plt.ylabel("Frequency")
plt.ylim(0, ymax)
plt.show()
```
