```python
import pandas as pd
```

```python
df_L = pd.read_csv("data/vad_delta_L.csv")
df_L.head()
```

```python
df_R = pd.read_csv("data/vad_delta_R.csv")
df_R.head()
```

```python
# データのサイズ
print(f"L(行, 列): {df_L.shape}")
print(f"R(行, 列): {df_R.shape}")
```

```python
# filename, text, delta類だけ抽出
df_L_delta = df_L[["filename", "text", "delta_valence", "delta_arousal", "delta_dominance", "delta_int_kushinada"]]
df_L_delta.to_csv("data/delta_L.csv", encoding="utf-8", index=False)
df_L_delta.head()
```

```python
# filename, text, delta類だけ抽出
df_R_delta = df_R[["filename", "text", "delta_valence", "delta_arousal", "delta_dominance", "delta_int_kushinada"]]
df_R_delta.to_csv("data/delta_R.csv", encoding="utf-8", index=False)
df_R_delta.head()
```

```python
# パーセンタイル
print("----------L---------")
print("<delta_arousal>")
for p in [80, 85, 90, 95]:
    print(p, df_L["delta_valence"].quantile(p/100).round(2))
print("")

print("<delta_int>")
for p in [80, 85, 90, 95]:
    print(p, df_L["delta_int_kushinada"].quantile(p/100).round(2))
print("")

print("----------R------------")
print("<delta_arousal>")
for p in [80, 85, 90, 95]:
    print(p, df_R["delta_valence"].quantile(p/100).round(2))
print("")

print("<delta_int>")
for p in [80, 85, 90, 95]:
    print(p, df_R["delta_int_kushinada"].quantile(p/100).round(2))
```
