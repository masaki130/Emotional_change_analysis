```python
import os
import glob
import re
import pandas as pd

trn_dir = "/autofs/diamond/share/corpus/CSJ/TRN/Form2/noncore"

# D03のF/Mのみ取得
files = glob.glob(os.path.join(trn_dir, "D03F*.trn")) + \
        glob.glob(os.path.join(trn_dir, "D03M*.trn"))

out_dir = "data/Scripts"
os.makedirs(out_dir, exist_ok=True)


for trn_file in files:
    rows = []

    with open(trn_file, "r", encoding="shift_JIS") as f:
        for line in f:

            line = line.strip()

            if not line:
                continue

            # 発話番号 時間 話者:内容
            match = re.match(
                r"(\d+)\s+(\d+\.\d+)-(\d+\.\d+)\s+(.):(.*)",
                line
            )

            if match:
                utt_id = match.group(1)
                start = float(match.group(2))
                end = float(match.group(3))
                speaker = match.group(4)
                text = match.group(5)

                rows.append([
                    utt_id,
                    start,
                    end,
                    speaker,
                    text
                ])

    df = pd.DataFrame(
        rows,
        columns=[
            "utt_id",
            "start",
            "end",
            "speaker",
            "text"
        ]
    )

    # ファイル名
    name = os.path.splitext(
        os.path.basename(trn_file)
    )[0]

    df.to_csv(
        f"{out_dir}/{name}.csv",
        index=False,
        encoding="utf-8"
    )

print("変換完了")
```
