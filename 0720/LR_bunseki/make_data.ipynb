```python
import pandas as pd

df_L = pd.read_csv("/home/mitani/CSJ-emo-int_bunseki/csj_vad_bunseki/speaker_split_noise/speaker_L_cleaned.csv")
df_R = pd.read_csv("/home/mitani/CSJ-emo-int_bunseki/csj_vad_bunseki/speaker_split_noise/speaker_R_cleaned.csv")
```

```python
df_L_vad = df_L[["filename", "text", "valence", "arousal", "dominance", "intensity_pred_kushinada"]]
df_L_vad.head()
```

```python
df_R_vad = df_R[["filename", "text", "valence", "arousal", "dominance", "intensity_pred_kushinada"]]
df_R_vad.head()
```

```python
# 隣接発話間の変化量を計算
df_L["delta_valence"] = df_L["valence"].diff().abs().round(2)       
df_L["delta_arousal"] = df_L["arousal"].diff().abs().round(2)
df_L["delta_dominance"] = df_L["dominance"].diff().abs().round(2)
df_L["delta_int_kushinada"] = df_L["intensity_pred_kushinada"].diff().abs().round(2)

df_R["delta_valence"] = df_R["valence"].diff().abs().round(2)
df_R["delta_arousal"] = df_R["arousal"].diff().abs().round(2)
df_R["delta_dominance"] = df_R["dominance"].diff().abs().round(2)
df_R["delta_int_kushinada"] = df_R["intensity_pred_kushinada"].diff().abs().round(2)

# 1行目は比較対象がないため削除(NaNになる)
df_diff_L = df_L.iloc[1:].copy()
df_diff_R = df_R.iloc[1:].copy()
```

```python
# filename, text, delta類だけ抽出
df_L_delta = df_L[["filename", "text", "delta_valence", "delta_arousal", "delta_dominance", "delta_int_kushinada"]]
df_L_delta.head()
```

```python
df_R_delta = df_R[["filename", "text", "delta_valence", "delta_arousal", "delta_dominance", "delta_int_kushinada"]]
df_R_delta.head()
```

```python
df_L.head()     # 確認(speaker_L_cleaned.csv + delta類)
```

```python
# 保存
df_L.to_csv("data/vad_delta_L.csv", encoding="utf-8", index=False)
df_R.to_csv("data/vad_delta_R.csv", encoding="utf-8", index=False)
```
