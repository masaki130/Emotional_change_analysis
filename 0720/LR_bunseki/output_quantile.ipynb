```python
# パーセンタイル毎にarousalの音響変化を伴う発話を抽出してcsvに保存
import pandas as pd

df_L = pd.read_csv("data/vad_delta_L.csv")
df_R = pd.read_csv("data/vad_delta_R.csv")
```

```python
L_80_a = df_L[df_L["delta_arousal"] > 0.29]
L_85_a = df_L[df_L["delta_arousal"] > 0.36]
L_90_a = df_L[df_L["delta_arousal"] > 0.45]
L_95_a = df_L[df_L["delta_arousal"] > 0.55]

R_80_a = df_R[df_R["delta_arousal"] > 0.28]
R_85_a = df_R[df_R["delta_arousal"] > 0.31]
R_90_a = df_R[df_R["delta_arousal"] > 0.35]
R_95_a = df_R[df_R["delta_arousal"] > 0.5]

L_80_int = df_L[df_L["delta_int_kushinada"] > 2.43]
L_85_int = df_L[df_L["delta_int_kushinada"] > 2.9]
L_90_int = df_L[df_L["delta_int_kushinada"] > 3.29]
L_95_int = df_L[df_L["delta_int_kushinada"] > 4.06]

R_80_int = df_R[df_R["delta_int_kushinada"] > 1.92]
R_85_int = df_R[df_R["delta_int_kushinada"] > 2.04]
R_90_int = df_R[df_R["delta_int_kushinada"] > 2.46]
R_95_int = df_R[df_R["delta_int_kushinada"] > 2.94]
```

```python
# 保存
import os

# 出力フォルダを作成
os.makedirs("results/L_a", exist_ok=True)                  
os.makedirs("results/R_a", exist_ok=True) 
os.makedirs("results/L_int", exist_ok=True) 
os.makedirs("results/R_int", exist_ok=True) 

# 抽出したパーセンタイルを保存
L_80_a.to_csv("results/L_a/L_80_a.csv", index=False)
L_85_a.to_csv("results/L_a/L_85_a.csv", index=False)
L_90_a.to_csv("results/L_a/L_90_a.csv", index=False)
L_95_a.to_csv("results/L_a/L_95_a.csv", index=False)

R_80_a.to_csv("results/R_a/R_80_a.csv", index=False)
R_85_a.to_csv("results/R_a/R_85_a.csv", index=False)
R_90_a.to_csv("results/R_a/R_90_a.csv", index=False)
R_95_a.to_csv("results/R_a/R_95_a.csv", index=False)

L_80_int.to_csv("results/L_int/L_80_int.csv", index=False)
L_85_int.to_csv("results/L_int/L_85_int.csv", index=False)
L_90_int.to_csv("results/L_int/L_90_int.csv", index=False)
L_95_int.to_csv("results/L_int/L_95_int.csv", index=False)

R_80_int.to_csv("results/R_int/R_80_int.csv", index=False)
R_85_int.to_csv("results/R_int/R_85_int.csv", index=False)
R_90_int.to_csv("results/R_int/R_90_int.csv", index=False)
R_95_int.to_csv("results/R_int/R_95_int.csv", index=False)
```
