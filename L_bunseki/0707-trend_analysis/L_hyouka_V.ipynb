```python
# LとRの音声で評価(今回は未使用)-str
import pandas as pd
import os
import shutil

# ==========================
# パス
# ==========================
l_csv = "/home/mitani/CSJ-emo-int_bunseki/0718/L_bunseki/vad_dis.csv"
lr_csv = "/home/mitani/CSJ-emo-int_bunseki/csj_vad_bunseki/speaker_split_noise/speaker_LR_merged.csv"
wav_dir = "/home/mitani/CSJ-emo-int_bunseki/ALL-WAV"
output_root = "/home/mitani/CSJ-emo-int_bunseki/0718/L_bunseki/0707-trend_analysis/L_eval-V-str_audio-LR"

os.makedirs(output_root, exist_ok=True)

# ==========================
# CSV読込
# ==========================
df_l = pd.read_csv(l_csv)
df_lr = pd.read_csv(lr_csv)

# L_vad_dis.csvで検出された行番号
change_idx = [
    5, 9, 16, 23, 28, 56, 58, 63, 66, 77, 128, 130, 144, 147, 160, 169, 186, 222,
    247, 249,
]

context = 3

# ==========================
# 抽出
# ==========================
for idx in change_idx:

    # L_vad_dis.csvのidx行目のファイル名
    target_filename = df_l.loc[idx, "filename"]

    print(f"{idx} -> {target_filename}")

    # speaker_LR_merged.csvから同じファイル名を検索
    matched = df_lr[df_lr["filename"] == target_filename]

    if matched.empty:
        print(f"見つかりません: {target_filename}")
        continue

    target_row = matched.index[0]

    # 前後context発話を抽出
    start = max(0, target_row - context)
    end = min(len(df_lr), target_row + context + 1)

    context_df = df_lr.iloc[start:end]

    # 保存フォルダ
    folder = os.path.join(
        output_root,
        os.path.splitext(target_filename)[0]
    )
    os.makedirs(folder, exist_ok=True)

    # context.csv保存
    context_df.to_csv(
        os.path.join(folder, "context.csv"),
        index=False,
        encoding="utf-8-sig"
    )

    # target情報保存
    with open(os.path.join(folder, "target.txt"), "w", encoding="utf-8") as f:
        f.write(f"L_vad_dis index : {idx}\n")
        f.write(f"Target filename : {target_filename}\n")
        f.write(f"speaker_LR_merged row : {target_row}\n")

    # 音声コピー
    for _, row in context_df.iterrows():

        src = os.path.join(wav_dir, row["filename"])

        if not os.path.exists(src):
            print(f"Not found: {src}")
            continue

        # TARGETを付ける
        if row["filename"] == target_filename:
            dst_name = "TARGET_" + row["filename"]
        else:
            dst_name = row["filename"]

        dst = os.path.join(folder, dst_name)

        shutil.copy2(src, dst)

print("完了")
```
